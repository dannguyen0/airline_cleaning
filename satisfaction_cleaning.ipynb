```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Reading the .csv file and opening a quick sample of what is represented in the dataset
df = pd.read_csv(r'C:\Users\Dan\Documents\Professional\Data Analytics\Airlines\airline_satisfaction.csv')

pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
nulls = df[df['Arrival Delay in Minutes'].isnull()]
nulls
```

```python
nulls['Arrival Delay in Minutes'].isnull().sum()
```

```python
# Copying the dataframe before dropping values to preserve original dataframe
ratings = df.copy().dropna()
ratings.info()
```

```python
ratings['Gender'].value_counts()
```

```python
delays_df = ratings[["satisfaction", "Departure Delay in Minutes"]]
delays_df
```

```python
plt.pie(ratings['satisfaction'].value_counts(), autopct = '%1.1f%%', labels = ['Satisfied', 'Dissatisfied'], colors = ['green', 'red'])
plt.show()
```

```python
correlations = ratings.corr(numeric_only = True)
correlations
```

```python
deepblue = sns.color_palette("light:#5A9", as_cmap=True)
sns.heatmap(data = correlations, annot = True, cmap = deepblue)
plt.rcParams['figure.figsize'] = (20, 20)
plt.show(block = True)
```

```python
# Changing satisfaction into numbers 0 - 1. 0 being disatisfied, 1 being satisfied. 

assignment = {'satisfied':1, 'dissatisfied':0} 
ratings['satisfaction'].replace(assignment, inplace = True)
ratings
```

```python
correlations = ratings.corr(numeric_only = True)
correlations
```

```python
deepblue = sns.color_palette("light:#5A9", as_cmap=True)
sns.heatmap(data = correlations, annot = True, cmap = deepblue)
plt.rcParams['figure.figsize'] = (20, 20)
plt.show(block = True)
```

```python
# Looking at satisfaction alone, it seems that there's too many factors that would influence a traveler's overall score. On the opposite ends, we see that the biggest factor leading to
# dissatisfaction are the delays. When there are delays, the satisfaction is even negative. Same can be said about gate location. 
```
